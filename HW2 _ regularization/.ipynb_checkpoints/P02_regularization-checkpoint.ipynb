- Regularization(정규화)

   
      정규화는 Overfitting 문제를 완화시키는 데 도움을 준다. Regression에서, 특정 Coefficient들이 다른 값들에 비해 상대적으로 매우 큰 값을 가지게 되면, 그 Model은 fluctuation이 크고, Overfitting 되었을 가능성이 크다. 그래서 Loss function에 additional term을 더하여 큰 값의 coefficient들의 크기를 panalize해주는 것이 정규화 과정이다. Ridge와 Lasso는 정규화의 종류 중 하나로 additional term으로 Ridge는 L2 norm, Lasso는 L1 norm을 사용하는 것이 특징이다. 각 additional term에는 Model의 complexity를 조절하는 hyper parameter(λ)가 존재한다. 이 λ가 커질수록, Coefficient들의 크기가 줄어들면서 High bias, Low variance를 가지게 된다. Ridge와 Lasso는 λ값이 변함에 따라 Coefficient들의 변하는 정도에 차이가 있는데, Ridge는 Coefficient의 크기가 줄지만 0이 되지는 않는 반면, Lasso는 특정 threshold이하로 Coefficient의 크기가 작아지면 아예 0으로 바꿔버린다. 그리고 이 threshold값은 λ에 의해 결정된다. 그래서 Lasso에서는 0의 값을 갖는 Coefficient들이 상대적으로 많이 존재하게 되면서 Model의 해석에 있어 좀 더 Coefficient과 Model간의 영향 관계를 더 쉽게 파악할 수 있게 된다.


- Coordinate Gradient Descent

    
      Coordinate Gradient Descent 방법은 최적화 알고리즘 중 하나로 반복적으로 각각의 좌표축을 따라 step하며 함수의 최솟값을 찾는 방법이다. 예를 들어 D+1개의 변수를 갖는 g(w0, w1, w2, ... , wD)함수가 있다고 가정하고, 이 함수의 Minimum point를 찾고 싶을 때, 각 변수 중 하나의 변수, 즉 하나의 좌표 축을 제외한 나머지의 좌표축은 고정시킨 후, 그 좌표 축 상의 Minimum을 갖는 값을 고정시키지 않은 변수 값에 넣는 것이다. 이 방법을 모든 변수에 대해 반복해 진행하다 보면 함수는 결국 모든 좌표축 방면에서 최소 값을 갖는 Minimum point에 도달하게 된다.

    - Weights(Values of weights, number of non-zero weights)   
    위의 마지막 그래프인 P2-4를 보면, λ값이 커질수록 non-zero weights의 수가 줄어드는 것을 확인할 수 있다. weights를 0으로 바꾸는 threshold의 범위가 0을 기준으로 [-λ/2, +λ/2]인데, λ값이 커진다는 것은 이 범위가 넓어진다는 뜻이니 0으로 바뀌는 weights들의 수가 늘어날 수 밖에 없다. weights의 값 또한 λ가 커질 수록 penalizing 정도가 커져 전반적으로 줄어든다. 위 P2-1의 결과 그래프를 보면 λ값이 클수록(왼쪽으로 갈수록) 0의 값을 가지는 weights들도 많아지고, 0이 되지 않은 weight들도 절대값은 점점 줄어드는 것을 확인 할 수 있다.
   
   
    - Training error
    Training error는 P2-2의 결과 그래프를 보면 λ값이 커질수록 error또한 커지는 것을 확인할 수 있는데, 이것은 λ가 Model의 Complexity를 조절하기 때문이다. λ가 클수록 Model의 Complexity는 줄어들면서 단순한 모델이 된다. 즉 bias는 커지고, variation은 줄어든다. 이럴때는 underfitting일 확률이 높다. training error는 overfitting일수록 값이 줄어드므로, 이 경우 training error의 값은 λ가 커질 수록 커지게 된다.
    
    
    - Test error

Model이 Overfitting 되어 있었으면, λ값이 커지면서 Model Complexity가 줄어들면서 Optimum point까지는 Test error가 줄어들다가 Underfitting 되기 시작하면서 다시 Test error의 값이 커져야 하는데, 이 그래프에서는 Test error값이 계속 커지는 모습만 보인다. 따라서 이 Model은 이미 underfitting 되어있었다는 결론을 내릴 수 있다. 최소 error값을 가져오는 최적의 λ값은 600/512 = 1.172 근처이거나, 혹은 이 값보다 더 작은 값일 것이라고 유추할 수 있다.
